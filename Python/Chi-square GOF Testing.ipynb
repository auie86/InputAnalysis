```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def chi2_gof(data: np.ndarray, num_bins: int, show:bool = True):
    # Calculate the expected frequency for each bin
    expected_frequency_per_bin = total_observations / num_bins

    # estimate the scale parameter from the data
    scale = np.mean(data)
    if show:
        print(f"Sample mean: {scale}")

    # Use the PPF (inverse CDF) to find the equi-probable bin boundaries
    # We want num_bins bins, so we need num_bins+1 boundaries. The probabilities are 0.0, 0.1, 0.2, ..., 1.0.
    # The 'loc' parameter is fixed to 0 for a standard exponential distribution.
    bin_probabilities = np.linspace(0, 1, num_bins + 1)
    bin_edges = stats.expon.ppf(bin_probabilities, scale=scale)

    # Calculate the observed frequencies
    observed_frequencies, _ = np.histogram(data, bins=bin_edges)

    # Create an array of expected frequencies (all the same)
    expected_frequencies = np.full(num_bins, expected_frequency_per_bin)

    # Perform the chi-square test
    chi2_stat, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies, ddof=1)

    if show:
        print(f"Chi-square statistic: {chi2_stat}")
        print(f"P-value: {p_value}")

    # Interpret the result
    alpha = 0.05
    if show:
        print(f"\nSignificance level (alpha): {alpha}")
        if p_value > alpha:
            print("The p-value is greater than alpha. We do not reject the null hypothesis.")
            print("The data is likely from an exponential distribution.")
        else:
            print("The p-value is less than or equal to alpha. We reject the null hypothesis.")
            print("The data is not likely from an exponential distribution.")
        print("\n")
    return scale, chi2_stat, p_value
```

```python
# Generate the same example data
np.random.seed(47)
scale = 125
num_obs = 750
data = stats.expon.rvs(scale=scale, size=num_obs)

# Number of bins
num_bins = 10
total_observations = len(data)

plt.hist(data)
```

```python
scale, chi2_stat, p_value = chi2_gof(data, 15, True)
print(f"scale: {scale}; chi2_stat: {chi2_stat}; p_value: {p_value}")
```
